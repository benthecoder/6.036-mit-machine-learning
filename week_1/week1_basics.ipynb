## Problem of induction
* The basis of learning from data
* why do we think that previously seen data -> predict future
* operationalize by making assumptions
  * all training data are IID and test are from same distribution 

### Solve two problems
1. Estimation - diff trials = diff results ? How to predict how well we estimate
2. Generalization - how can we predict results on data we've never encountered before

### 6 Characteristics for problem and solution
 
1. Problem class - Nature of train data and queries made at testing time
2. Assumptions - source of data and form of solution
3. Evaluation criteria - goal of prediction, answer evaluations, overall performance of system
4. Model Type - intermediate model? aspects of data modelled? How is it used to make predictions
5. Model Class - what parametric class to use, what criteria to pick model
6. Algorithm - what computational process used to fit model to data or to make predictions

Without assumptions about nature of process generating data, we can't perform generalization.

## Assumptions

1. data are IID
2. data generated by Markov Chains
3. process generating data is adversarial
4. 'True' model generating data can be perfectly described by a set of hypothesis

These assumptions are to reduce the size of possible space of solutions to reduce the need for more data


## Evaluation predictions

To evaluate performace, we need a loss function, L(g, a) that tells us how bad it would be to make a guess g, when a is the right answer. A few types of loss functions are 

1. 0-1 loss
2. Squared loss (MSE)
3. Linear loss (ASE)
4. Assymetric loss (when wrong guess is more expensive)

## Prediction Rule

1. Fit the model to training data
2. use model to make predictions

> Write objective function (training error), come up with hypothesis class, use computational optimization methods, find best elements in hypothesis class according to objective function

## Model class and parameter fitting

Parametric - fixed, finite amount of parameters
Non-parametric - does not assume structure of data

Model selection - pick model class M from set of possible model classes
Model fitting - pick model within that class specified by parameters 

## Algorithm

After describing class of models and ways of scoring them, we have to find algorithms that can help us determine parameters theta that minimizes E(theta)

## Classification
binary classifier maps Rd -> {-1, 1}. we use H to stand for hypotehsis. The classification process looks like this: 

x -> h -> y  

Classifiers that work well on new data is useful, think of it like in final exames, where the student has to generalize based on problem sets in class. Train and test data are drawn independantly from same prob distribution.

## Learning algorithm

Hypothesis class H is set of finite or infinite of possible classifiers

Learning algorithm is procedure that takes Data set Dn as input, and returns element h of H

Dn -> learning alg (H) -> h

find choice of H that minimizes test error of h. one way to do that is to restrict size of H 

## Evaluating learning algorithm

Best method is to measure test error on data

1. train on new training set
2. evaluate resulting h (learned hypothesis) on testing set that doesn't overlap the training set

### Cross-validation
